### Does the size (diameter) and relative velocity of an asteroid influence how close it comes to Earth (miss distance)?

I want to use linear regression to analyze if larger or faster-moving asteroids tend to pass closer to Earth. This question would aid in my understanding  of how these factors affect the likelihood of an asteroid being hazardous.





```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv("neo.csv")

ind = ["relative_velocity", "est_diameter_min"]
dep = "miss_distance"

model = LinearRegression()
model.fit(df[ind], df[dep])  

df["Predicted Miss Distance"] = model.predict(df[ind])
df
```

The linear regression model predicts miss distance based on two independent variables: relative velocity and estimated minimum diameter.
Relative Velocity: Higher velocities may lead to smaller miss distances, suggesting that faster-moving asteroids could pass closer to Earth. This trend is supported by the physical nature of fast-moving objects.
Estimated Minimum Diameter: Larger objects may influence the predicted miss distance, with larger asteroids potentially having a larger gravitational effect or a larger space around them, which could either bring them closer or push them farther, depending on other factors (such as their orbit).
